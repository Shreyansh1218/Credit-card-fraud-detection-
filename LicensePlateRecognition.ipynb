```python
# Install necessary libraries
!pip install easyocr opencv-python matplotlib

import cv2
import easyocr
import matplotlib.pyplot as plt
from google.colab import files

# Upload an image
uploaded = files.upload()

# Load the first uploaded image
image_path = list(uploaded.keys())[0]

# Read the image
img = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])

# Perform OCR on the entire image
results = reader.readtext(gray)

# Draw rectangles and labels on detected text
for (bbox, text, prob) in results:
    top_left = (int(bbox[0][0]), int(bbox[0][1]))
    bottom_right = (int(bbox[2][0]), int(bbox[2][1]))
    cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(img, text, (top_left[0], top_left[1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0), 2)

# Display the image with OpenCV and Matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Print detected texts
detected_texts = [res[1] for res in results]
print("Detected text(s):", detected_texts)
```
